```python
# image read
%matplotlib inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
img_in = cv2.imread('test1.jpg')
plt.figure(1)
plt.title('Input Image')
plt.imshow( cv2.cvtColor( img_in , cv2.COLOR_BGR2RGB ) )
```

```python
# RGB to Gray Conversion
img_gray = cv2.cvtColor( img_in , cv2.COLOR_BGR2GRAY )
plt.figure(2)
plt.title('Grayscale image')
plt.imshow( img_gray , cmap = plt.get_cmap('gray') , vmin=0 , vmax=255 )
```

```python
# Histogram Equalization
img_hist = cv2.equalizeHist( img_gray )
plt.figure(3)
plt.title('Histogram Equalization')
plt.imshow( img_hist , cmap = plt.get_cmap('gray') , vmin=0 , vmax=255)
```

```python
#Face Detection
face_haar_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
img_tmp = img_in.copy()
all_faces = face_haar_cascade.detectMultiScale( img_hist , 1.9)
for (x,y,w,h) in all_faces:
    cv2.rectangle(img_tmp,(x,y),(x+w,y+h),( np.random.random()*255 , np.random.random()*255 , np.random.random()*255 ) , 5 )
plt.figure(4)
plt.title('Faces with rectangle box')
plt.imshow( cv2.cvtColor(img_tmp , cv2.COLOR_BGR2RGB) ) 
```

```python
#Face Croped from Histogram Equalized image and resized to 100*100
fig_no = 4
for (x,y,w,h) in all_faces:
    face = img_in[ y:y+h , x:x+w ]
    face_resized = cv2.resize( face , (1000,1000) , interpolation = cv2.INTER_AREA )
    plt.figure( fig_no )
    plt.title( 'Face ' + str(fig_no - 3) )
    plt.imshow( cv2.cvtColor( face_resized , cv2.COLOR_BGR2RGB ) , cmap = plt.get_cmap('gray') , vmin=0 , vmax=255 )
    fig_no = fig_no+1
```
